```python
# get the the mean values and std for each model
def get_mean_std(data):
    data = np.vstack(data)
    mean, std = np.mean(data, axis=0), np.std(data, axis=0)
    return mean.round(2), std.round(2)
```

```python
# K=1
mobile_zero_1_mean, mobile_zero_1_std = get_mean_std([mobile_zero_1_1,
                                                      mobile_zero_1_2,
                                                      mobile_zero_1_3,
                                                      mobile_zero_1_4,
                                                      mobile_zero_1_5,
                                                      mobile_zero_1_6]
                                                     )
mobile_nshot_1_mean, mobile_nshot_1_std = get_mean_std([mobile_nshot_1_1,
                                                         mobile_nshot_1_2,
                                                         mobile_nshot_1_3,
                                                         mobile_nshot_1_4,
                                                         mobile_nshot_1_5,
                                                         mobile_nshot_1_6]
                                                        )
cnn_zero_1_mean, cnn_zero_1_std = get_mean_std([cnn_zero_1_1,
                                                cnn_zero_1_2,
                                                cnn_zero_1_3,
                                                cnn_zero_1_4,
                                                cnn_zero_1_5,
                                                cnn_zero_1_6]
                                               )
cnn_nshot_1_mean, cnn_nshot_1_std = get_mean_std([cnn_nshot_1_1,
                                                   cnn_nshot_1_2,
                                                   cnn_nshot_1_3,
                                                   cnn_nshot_1_4,
                                                   cnn_nshot_1_5,
                                                   cnn_nshot_1_6]
                                                  ) 

# k=2
mobile_zero_2_mean, mobile_zero_2_std = get_mean_std([mobile_zero_2_1,
                                                      mobile_zero_2_2,
                                                      mobile_zero_2_3
                                                      ])
mobile_nshot_2_mean, mobile_nshot_2_std = get_mean_std([mobile_nshot_2_1,
                                                         mobile_nshot_2_2,
                                                         mobile_nshot_2_3
                                                         ])
cnn_zero_2_mean, cnn_zero_2_std = get_mean_std([cnn_zero_2_1,
                                                cnn_zero_2_2,
                                                cnn_zero_2_3
                                                ])
cnn_nshot_2_mean, cnn_nshot_2_std = get_mean_std([cnn_nshot_2_1,
                                                   cnn_nshot_2_2,
                                                   cnn_nshot_2_3
                                                   ]) 

# k=3
mobile_zero_3_mean, mobile_zero_3_std = get_mean_std([mobile_zero_3_1,
                                                      mobile_zero_3_2
                                                      ])
mobile_nshot_3_mean, mobile_nshot_3_std = get_mean_std([mobile_nshot_3_1,
                                                         mobile_nshot_3_2
                                                         ])
cnn_zero_3_mean, cnn_zero_3_std = get_mean_std([cnn_zero_3_1,
                                                cnn_zero_3_2
                                                ])
cnn_nshot_3_mean, cnn_nshot_3_std = get_mean_std([cnn_nshot_3_1,
                                                   cnn_nshot_3_2
                                                   ])
```

```python
# get the statics
fold_1_mean = np.vstack([cnn_zero_1_mean,
                         mobile_zero_1_mean,
                         cnn_nshot_1_mean,
                         mobile_nshot_1_mean])
fold_1_std = np.vstack([cnn_zero_1_std,
                         mobile_zero_1_std,
                         cnn_nshot_1_std,
                         mobile_nshot_1_mean])

fold_2_mean = np.vstack([cnn_zero_2_mean,
                         mobile_zero_2_mean,
                         cnn_nshot_2_mean,
                         mobile_nshot_2_mean])
fold_2_std = np.vstack([cnn_zero_2_std,
                         mobile_zero_2_std,
                         cnn_nshot_2_std,
                         mobile_nshot_2_mean])

fold_3_mean = np.vstack([cnn_zero_3_mean,
                         mobile_zero_3_mean,
                         cnn_nshot_3_mean,
                         mobile_nshot_3_mean])
fold_3_std = np.vstack([cnn_zero_3_std,
                         mobile_zero_3_std,
                         cnn_nshot_3_std,
                         mobile_nshot_3_mean])
```

```python
## Plot precision
matplotlib.rcParams.update({'font.size': 13})

models = ("1-shot learning", "2-shot learning", "3-shot learning")
penguin_means = {
    'Shallow CNN w/o refinement': (fold_1_mean[0, 0], fold_2_mean[0, 0], fold_3_mean[0, 0]),
    'MobileNet w/o refinement': (fold_1_mean[1, 0], fold_2_mean[1, 0], fold_3_mean[1, 0]),
    'CNN with refinement': (fold_1_mean[2, 0], fold_2_mean[2, 0], fold_3_mean[2, 0]),
    'MobileNet with refinement': (fold_1_mean[3, 0], fold_2_mean[3, 0], fold_3_mean[3, 0]),
}

x = np.arange(len(models))  # the label locations
width = 0.22  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')
colors = ["white", "c", "white", "c"]
edgecolors = ["black", "black", "black", "black"]
hatchs = ["/", "\\", "", ""]

for i, (attribute, measurement) in enumerate(penguin_means.items()):
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute, color=colors[i], edgecolor=edgecolors[i], hatch=hatchs[i])
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Precision')
ax.set_xticks(x + width, models)
ax.legend(loc='upper left', ncols=1)
ax.set_ylim(0, 1)

results_dir = Path(f'./optimized_models/refine_model/plot_evluations/')
results_dir.mkdir(exist_ok=True)
plt.savefig(os.path.join(results_dir, 'Precision.png'))

plt.show()
```
